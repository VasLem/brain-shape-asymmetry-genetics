```python
import os
import pandas as pd
with open('../SAMPLE_DATA/mosquito.csv', 'r') as inp:
    original = pd.read_csv(inp)
```

```python
original['id'] = original['index'].apply(lambda x: int(x.split('_')[1]))
original['side'] = original['index'].apply(lambda x: x.split('_')[2])
original['rep'] = original['index'].apply(lambda x: x.split('_')[3])
```

```python
for t in range(1, 3):
    augmented = original.copy()
    for l in range(t):            
        augmented_id = original['id'] + max(augmented['id'])
        to_add = original.copy()
        to_add['id'] = augmented_id
        augmented = pd.concat([augmented, to_add])
    augmented['index'] = augmented['id'].astype(str).apply(lambda x: '{0:0>2}'.format(x)) + "_" + augmented['side'] + '_' + augmented['rep']
    augmented.drop(columns=['id','side','rep'],inplace=True)
    augmented.to_csv(f'../SAMPLE_DATA/augmented_mosquito_reps_{t}.csv', index=False)
```

```python
for t in [5,8]:
    reduced = original[original['id']<=t].copy()
    reduced.drop(columns=['id','side','rep'],inplace=True)
    reduced.to_csv(f'../SAMPLE_DATA/reduced_mosquito_{t}.csv', index=False)
```
